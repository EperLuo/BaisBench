```python
import os
import yaml
```

```python
all_h5ad = [file for file in os.listdir('../Task1_data') if file.endswith('raw.h5ad')]
```

```python
for h5ad in all_h5ad:
    organ = h5ad.replace('_raw.h5ad','')
    os.makedirs(f'/stor/lep/workspace/AISbenchmark/AutoBA/examples/{organ}', exist_ok=True)
    data_list = [f'/stor/lep/workspace/AISbenchmark/Task1_data/{h5ad}: single-cell RNA-seq data from human {organ}.']
    output_dir = f'/stor/lep/workspace/AISbenchmark/AutoBA/examples/{organ}/output'
    goal_description = f'Given this single-cell RNA-seq dataset from human {organ}, annotate the cell type for each cell and return the h5ad file to me. Annotation must be done based on knowledge; external tools like CellTypist cannot be used'
    config = {
        'data_list': data_list,
        'output_dir': output_dir,
        'goal_description': goal_description
    }
    with open(f'./examples/{organ}/config.yaml', 'w') as f:
        yaml.dump(config, f)
```
